# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['diagnosis']=le.fit_transform(df['diagnosis'])
```

```python
df['diagnosis'].value_counts()
```

```python
df=df.drop('Unnamed: 32',1)
```

# Feature Selection

```python
 import seaborn as sns
```

```python
correlations=df.corr()
top_correlations=correlations.index

plt.figure(figsize=(24,24))
sns.heatmap(data=df[top_correlations].corr(),annot=True,cmap='RdYlGn')
```

```python
X=df.drop(['diagnosis'],1)
Y=df['diagnosis']
```

```python
from sklearn.tree import ExtraTreeClassifier
```

```python
ET=ExtraTreeClassifier()
```

```python
ET.fit(X,Y)
```

```python
feature_importances=pd.Series(ET.feature_importances_,index=X.columns)
feature_importances
```

# Creating test and train data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X[feature_importances.index],Y,random_state=2,test_size=0.2)
```

# Scaling the features

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
X_train=sc.fit_transform(X_train)
```

```python
X_test=sc.fit_transform(X_test)
```

# Hyper Parameter Tuning

```python
params={'penalty':['l2',None],
       'C': [1,2],
       'class_weight': ['balanced',None]}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score
```

```python
lr=LogisticRegression()
```

```python
gs=GridSearchCV(estimator=lr,param_grid=params,cv=5,scoring=precision_score)
```

# Model Building and Evaluation

```python
gs.fit(X_train,Y_train)
```

```python
gs.best_params_
```

```python
final_lr_model=LogisticRegression(C=1, class_weight='balanced', penalty='l2')
```

```python
final_lr_model.fit(X_train,Y_train)
```

```python
predict=final_lr_model.predict(X_test)
```

```python
precision_score(predict,Y_test)
```

# Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
parameters={'n_estimators':[100,200,300,500],
            'criterion': ['gini','entropy'],
            'max_depth':[1,2,4,'None'],
            'max_leaf_nodes':[2,'None'],
            'max_features':['auto','sqrt'] }
```

```python
gv=GridSearchCV(estimator=rf,param_grid=parameters,cv=5,scoring=precision_score)
```

```python
gv.fit(X_train,Y_train)
```

```python
gv.best_params_
```

```python
final_rf_model=RandomForestClassifier(criterion='gini',max_depth=1,max_features='auto',max_leaf_nodes=2,n_estimators=100)
```

```python
final_rf_model.fit(X_train,Y_train)
```

```python
rf_predict=final_rf_model.predict(X_test)
```

```python
precision_score(rf_predict,Y_test)
```
